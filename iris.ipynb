```python
from sklearn.datasets import load_iris
iris = load_iris()
print(type(iris))

```

```python
print(iris.keys())
```

```python
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data
```

```python
data.shape    #(150, 5)
```

```python
data.head()
```

```python
data.plot(kind="scatter", x="sepal length (cm)", y="sepal width (cm)") #matplotlib 사용
sns.jointplot(x="sepal length (cm)", y="sepal width (cm)", data=data, size=5) #sns 사용
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
#splliting data for testing and training
x_train, x_test, y_train, y_test = train_test_split(iris['data'], iris['target'])
#initializing knn
knn = KNeighborsClassifier(n_neighbors=5)
#training
knn.fit(x_train, y_train)
#testing
predicted_specis = knn.predict(x_test)
print("predicted data\n{}".format(predicted_specis))  ##\n 밑에 칸에…
#printing % of accuracy using mean
print("accuracy: {:.2f}%".format(100*np.mean(predicted_specis == y_test))) 
##.2f 2 decimal point 
```
